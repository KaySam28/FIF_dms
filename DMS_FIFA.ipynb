```python
##dms.shape gave us 4572 rows and dms.tai( is giving us 3720 so we can make the assumption that there are null below and as such can use the iloc to find out at excaty which point our data starts to turn null and delete it
dms.iloc[850:900]
```

```python
dms.drop(range(852,4572), inplace=True)
dms.shape
```

```python
##attendance null values
dms.Attendance.sort_values(ascending=True)
```

```python
##They occur in rows 823 and 841. I want to look for similarities in these two rows and try to draw a conclusion about why the data is missing
dms.iloc[700:900]['Stadium'].value_counts()
```

```python
## The stadium with null value twice (Estadio Beira-Rio) has 6games played there but has only two attendance as null so we tried to fill the null values wiht the average attendance 
avg_attendance = dms.Attendance.dropna().sum() / len(dms.Attendance.dropna())
print(avg_attendance)

dms.Attendance = dms.Attendance.fillna(avg_attendance)
dms.isnull().sum()
```

```python
## Data types within our data
dms.dtypes
```

```python
##Goals, attendance and year are in floats and we want to change that to integars
dms['Home Team Goals'] = dms['Home Team Goals'].astype('int64')
dms['Away Team Goals'] = dms['Away Team Goals'].astype('int64')
dms['Half-time Home Goals'] = dms['Half-time Home Goals'].astype('int64')
dms['Half-time Away Goals'] = dms['Half-time Away Goals'].astype('int64')
dms['Attendance'] = dms['Attendance'].astype('int64')
dms.dtypes
```

```python
##converting datetime from object to datetime data format
dms.Datetime = pd.to_datetime(dms.Datetime)
dms.dtypes
```

```python
##To start our analysis we have to plot so import maplotlib
import matplotlib.pyplot as mds
```

```python
goals_by_home_team = dms[['Home Team Goals','Home Team Initials']].groupby('Home Team Initials')
ax = goals_by_home_team.sum() \
.sort_values('Home Team Goals', ascending = False) \
.head(25) \
.plot(kind ='bar', figsize =(10,8))

goals_by_away_team = dms[['Away Team Goals','Away Team Initials']].groupby('Away Team Initials')
goals_by_away_team.sum() \
.sort_values('Away Team Goals', ascending = False) \
.head(25) \
.plot(kind ='bar', figsize =(10,8), ax=ax,stacked = True, color ='red')
mds.title('Total Goals Scored by Team')
mds.xlabel('Team Initials')
mds.ylabel('# of Goals');
```

```python
dms['Total Goals'] = dms['Home Team Goals'] + dms['Away Team Goals']
dms.head()
```

```python
grouped_by_year = dms[['Total Goals','Year']].groupby('Year')
goals_per_year = grouped_by_year['Total Goals'].sum()
goals_per_year.plot(color = 'red', figsize =(10,7))
mds.title('Total Goals Scored', fontsize = 18)
mds.xlabel('World Cup')
mds.ylabel('# of Goals');
```

```python
games_per_year = dms['Year'].value_counts()
games_per_year.sort_index() \
.plot(color = 'red', figsize = (10,7));
mds.title('Total Games Played', fontsize = 18)
mds.xlabel('World Cup')
mds.ylabel('# of Games');
```

```python
##Goals PER game
goals_per_game_per_year = goals_per_year / games_per_year
goals_per_game_per_year.plot(figsize = (10,7), color = 'Teal')
mds.title('Average Goals per Game by Year', fontsize = 18)
mds.xlabel('World Cup')
mds.ylabel('Average # of Goals per Game');
```
